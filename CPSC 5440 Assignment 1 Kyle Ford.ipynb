```python
#Open the English file
english_file = open("english.txt")
```

```python
#read each line as its own list element. 
english_lines = english_file.readlines()

#print first 10 elements to show accuracy
print(english_lines[:10])
```

```python
#define training and target datasets
training_dataset = []
target_dataset = []

for word in english_lines:
    #remove \n character
    cleaned_word = word.replace("\n", '')
    
    #check if word is 5 characters
    if len(cleaned_word) == 5:
        #array for converting word to ord representation
        word_to_ord = []
        
        #iterate through cleaned word characters, ord the characters, and append it to the word_to_ord list
        for char in cleaned_word: 
            word_to_ord.append(ord(char))
            
        #append teh ord'ed word to the training dataset
        training_dataset.append(word_to_ord)
        
        #append the correct answer to the target dataset
        target_dataset.append(0)
```

```python
#print first 10 in training data
print("First 10 in training data: ")
for word in training_dataset[:10]: print(word)

#print first 10 in target data
print("First 10 in target data: ")
print(f"\nFirst 10 in target data: \n{target_dataset[:10]}")
```

```python
for word in french_lines:
    #remove \n character
    cleaned_word = word.replace("\n", '')
    
    #check if word is 5 characters
    if len(cleaned_word) == 5:
        #array for converting word to ord representation
        word_to_ord = []
        
        #iterate through cleaned word characters, ord the characters, and append it to the word_to_ord list
        for char in cleaned_word: 
            word_to_ord.append(ord(char))
            
        #append teh ord'ed word to the training dataset
        training_dataset.append(word_to_ord)
        
        #append the correct answer to the target dataset
        target_dataset.append(2)
```

```python
#import libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.neural_network import MLPClassifier

#create KNN, SVM, and MLP objects
knn_model = KNeighborsClassifier()
svm_model = svm.SVC()
mlp_nn = MLPClassifier()
```

```python
#split dataset into training and testing data
from sklearn.model_selection import train_test_split

#split into training and testing data. we use a 75/25 split
X_train, X_test, y_train, y_test = train_test_split(training_dataset, target_dataset, test_size = 0.25, random_state = 16)
```

```python
#fit each model using the training and target datasets
knn_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
mlp_nn.fit(X_train, y_train)
```

```python
from sklearn import metrics

#predict using each model and the testing dataset
knn_predicted = knn_model.predict(X_test)
svm_predicted = svm_model.predict(X_test)
mlp_nn_predicted = mlp_nn.predict(X_test)

#measure accuracy of each model
print("KNN accuracy:", metrics.accuracy_score(y_test, knn_predicted))
print("SVM accuracy: ", metrics.accuracy_score(y_test, svm_predicted))
print("MLP accuracy: ", metrics.accuracy_score(y_test, mlp_nn_predicted))
```

```python
#graph the accuracy results. We achieve 67%, 69%, and 68% accuracy for each model, respectively.
import numpy as np
import matplotlib.pyplot as plt

labels = ("KNN", "SVM", "MLP")

value = [67, 69, 68]

plt.title("Model Accuracy")

plt.xlabel("Accuracy")

y_pos = np.arange(len(labels))
plt.barh(y_pos, value, align = 'center', alpha = 0.5)
plt.yticks(y_pos, labels)

plt.show()
```
